1) Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при **alpha = 5%**. Дайте краткие рекомендации команде.

Результаты: 

1) Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550 . 

2) Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228 

3) Конверсии : conv1 = 1.05% conv2 = 1.47% .

 Сформулируем гипотезы: z-критерий для разности долей.

 **Нулевая гипотеза** – между конверсией в группах нет статистически значимых различий

 **Альтернативная гипотеза** – между конверсией в группах есть статистически значимые различия

2. Уровень статистической значимости α = 0.05.  

3. Непараметрические не используют предположения о распределении, а оперируют рангами и частотами.



4. Для оценки статистической значимости различий между конверсиями в двух группах можно использовать z-критерий для разности долей. 

 **Вычисляем стандартную ошибку разности долей:**

$$SE = \sqrt{\frac{conv_1*(1-conv_1)}{n_1} + \frac{conv_2*(1-conv_2)}{n_2}}$$


  **Вычисляем z-статистику:**

$$z = \frac{(conv_1 - conv_2)}{SE} $$

```python
conv1, conv2 = 1.05 / 100, 1.47/ 100
n1, n2 = 15550, 15550
SE = (conv1 *(1 - conv1) / n1 + conv2 *(1 - conv2) / n2)**0.5
z = (conv1 - conv2) / SE
print(f'SE  = {SE}')
print(f'z-критерий  = {-z}')
```

**Ответ** Для alpha = 5% критическое значение z-статистики для двусторонней критической области равно 1.96. Таким образом, полученное значение z-статистики сильно выше критического значения, что говорит о том, что **существует статистически значимая разница в конверсиях между группами.**

**Краткая рекомендация команде:** в этом эксперименте группа с упрощенной формой заказа показала более высокую конверсию в покупку, чем контрольная группа. # Задача 2
Сравниваем метрику конверсия в покупку. Размер выборки - 10000 элементов в каждой группе . Какой статистический критерий тут лучше всего подойдёт и почему ?

**Ответ** 

Для сравнения метрики конверсии в покупку на двух группах, лучше всего использовать статистический критерий, основанный на распределении Бернулли. Таким критерием является Z-тест. 

Z-тест используется для проверки гипотезы о равенстве долей в двух выборках. Он позволяет выявить, является ли разница между долями статистически значимой или случайной.

Для применения этого критерия необходимо знать общее количество наблюдений, количество успехов (т.е. покупок) и пропорцию (долю) успехов в каждой группе.

Z-тест подходит для выборок большого объема, что в данном случае (каждая выборка содержит 10 000 элементов) является достаточным условием для его применения. 

Поэтому, Z-тест является лучшим вариантом статистического критерия для сравнения метрики конверсии в покупку в данной задаче.
